```python
annotation_index_data=final['annotation'][::]
```

```python

def json_data():
    start=[]        
    end=[]
    label=[]
    text=[]
    for data in annotation_index_data:
        start.append(data['start'])
        end.append(data['end'])
        label.append(data['label'])
        text.append(data['text'])
    final['start']=start    
    final['end']=end
    final['label']=label
    final['text']=text
json_data()
```

```python
final.drop(['annotation'],axis=1,inplace=True)
```
