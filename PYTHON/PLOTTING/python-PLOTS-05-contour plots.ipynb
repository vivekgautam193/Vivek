## Contour Plots

A contour line or isoline of a function of two variables is a curve along which the function has a constant value.

It is a cross-section of the three-dimensional graph of the function f(x, y) parallel to the x, y plane.

Contour lines are used e.g. in geography and meteorology.

In cartography, a contour line joins points of equal elevation (height) above a given level, such as mean sea level.

We can also say in a more general way that a contour line of a function with two variables is a curve which connects points with the same values.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
n, m = 7, 7
start = -3

x_vals = np.arange(start, start+n, 1)
y_vals = np.arange(start, start+m, 1)

X, Y   = np.meshgrid(x_vals, y_vals)

print(X)
print(Y)
```

```python
fig, ax = plt.subplots()

ax.scatter(X, Y, color="green")
ax.set_title('Regular Grid, created by Meshgrid')
ax.set_xlabel('x')
ax.set_ylabel('y');
```

```python
import numpy as np

xlist = np.linspace(-3.0, 3.0, 400)
ylist = np.linspace(-3.0, 3.0, 400)

X, Y = np.meshgrid(xlist, ylist)
```

```python
Z = np.sqrt(X**2 + Y**2)
```

#### 3D plot surface, contour plots

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig, (ax) = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

ax[0] = fig.add_subplot(1,2,1, projection='3d')

ax[0].plot_surface(X, Y, Z)

ax[0].set_title('3D surface of X, Y and Z')
ax[0].set_xlabel('X ')
ax[0].set_ylabel('Y ')
ax[0].set_zlabel('Z ');

cp = ax[1].contour(X, Y, Z)

ax[1].clabel(cp, inline=True, fontsize=10)
ax[1].set_title('Contour Plot')
ax[1].set_xlabel('X ')
ax[1].set_ylabel('Y ');
```

```python
plt.figure(figsize=(10, 6))

contour = plt.contourf(X, Y, Z)
plt.colorbar(contour)

plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)');
```

```python
plt.figure(figsize=(10, 6))

contour = plt.contourf(X, Y, Z)
plt.clabel(contour, colors = 'k', fmt = '%2.1f', fontsize=12)
c = ('#ff0000', '#ffff00', '#0000FF', '0.6', 'c', 'm')
contour_filled = plt.contourf(X, Y, Z, colors=c)
plt.colorbar(contour)

plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)');
```

#### Levels

can be defined manually, by providing a list of levels as a fourth parameter. Contour lines will be drawn for each value in the list, if we use contour. 

For contourf, there will be filled colored regions between the values in the list.

```python
plt.figure(figsize=(10, 6))

levels = [0.0, 0.2, 0.5, 0.9, 1.5, 2.5, 3.5]
levels = np.linspace(0, 3.5, 10)
contour = plt.contourf(X, Y, Z, levels=levels)
plt.colorbar(contour)

plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)');
```
