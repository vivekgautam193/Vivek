#### Calculating Magnitude —

```python
import math
v = np.array([2, 3])

vMag = math.sqrt(v[0]**2 + v[1]**2)

print ('Magnitude of {} vector is = {}'.format(v, vMag))
```

```python
# By using linalg library provided by numpy
v = np.array([2,3])
vMag = np.linalg.norm(v)
print (vMag)
```

#### Scalar Multiplication - 
scalar multiplication — multiplying a vector by a single numeric value.

```python
v = np.array([-1,1])
w = 3 * v

print(w)

vecs   = np.array([v, w])

# Plot w
origin = [0], [0]
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:, 0], vecs[:, 1], color=['r', 'b'], scale=10)
```

#### Dot product -

multiplies each component pair of the vectors and sums the results. To indicate that we are performing a dot product operation, we use the • operator:

```python
x = np.array([2, 4])
y = np.array([1, 3])
```

The dot product of the above two vectors is (2 x 1) + (4 x 3) = 14.

```python
dot_product = 0

for a,b in zip(x,y):
    dot_product += a * b

print(dot_product)
```

```python
a = x * y
print(a.sum())
```

```python
print(x.dot(y))
```

#### The Cosine Rule -

measures the similarity between two vectors of an inner product space. 

It is measured by the cosine of the angle between two vectors and determines whether two vectors are pointing in roughly the same direction. 

It is often used to measure document similarity in text analysis.

In text analysis, a document can be represented by thousands of attributes/columns/features, each representing the frequency of a particular word. 

Thus, each document is an object represented by what is called a __term-frequency vector__. 

For example, Document1 contains five instances of the word team, while hockey occurs three times. The word coach is absent from the entire document, as indicated by a count value of 0. Such data can be highly asymmetric.



Term-frequency vectors are typically very long and sparse (i.e., they have many 0 values). 

Applications using such structures include 
- information retrieval, 
- text document clustering, 
- biological taxonomy, and 
- gene feature mapping. 

For example, two term-frequency vectors may have many 0 values in common, meaning that the corresponding documents do not share many words, but this does not make them similar. 

We need a measure that will focus on the words that the two documents do have in common, and the occurrence frequency of such words. 

In other words, we need a measure for numeric data that ignores zero-matches.

__Cosine similarity__ is a measure of similarity that can be used to compare documents or, say, give a ranking of documents with respect to a given vector of query words. 

Let x and y be 2 vectors for comparison. 

Using the __cosine__ measure as a similarity function, we have

$$ \Large similarity =  \frac{x \bullet y}{ \sqrt{x \bullet x} \sqrt{y \bullet y}} $$

```python
a  = [3, 45, 7,   2]
b  = [2, 54, 13, 15]
```

```python
cos_sim = np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))
cos_sim
```

#### some detailing about quiver plots ...

A quiver plot displays the vectors as arrows with components (u,v) at the points (x, y).

quiver(x, y, u, v)

__Parameters__
The following table lists down the different parameters for the Quiver plot −

    x	1D or 2D array, sequence. The x coordinates of the arrow locations
    y	1D or 2D array, sequence. The y coordinates of the arrow locations
    u	1D or 2D array, sequence. The x components of the arrow vectors
    v	1D or 2D array, sequence. The y components of the arrow vectors
    c	1D or 2D array, sequence. The arrow colors

style	[ 'sci' (or 'scientific') | 'plain' ] plain turns off scientific notation

If you have the vector field U and V, you can normalize them by:

N = numpy.sqrt(U^2 + V^2)  # there may be a faster numpy "normalize" function

U2, V2 = U/N, V/N

Then you can apply whatever scaling factor array you want:

U2 *= F

V2 *= F

```python
fig, ax = plt.subplots(figsize=(5, 5))

X = np.array((0))
Y = np.array((0))
U = np.array((-2))
V = np.array((-2))

ax.grid()
ax.quiver(X, Y, U, V, units='xy', scale=1)

ax.set_title('Quiver plot with one arrow')
ax.set_aspect('equal')

plt.xlim(-6, 6)
plt.ylim(-6, 6);
```

```python
N=np.sqrt(U**2 + V**2)
N
```

```python
U2, V2 = U/N, V/N
U2, V2
```

```python
scale = 1
```

```python
U2_scale = U2 * scale
V2_scale = V2 * scale

U2_scale, V2_scale
```
