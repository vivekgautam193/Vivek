## Converting the datatype of cols from object to int or float as per the requirement.

```python
hiv_main_data['Number of new HIV infections']=hiv_main_data['Number of new HIV infections'].astype(int)
```

```python
#separating 500 from &lt;500 in the col named Number of deaths due to HIV/AIDS.
hiv_main_data['Number of deaths due to HIV/AIDS']=hiv_main_data['Number of deaths due to HIV/AIDS'].str.extract(r'(\d+(?:\.\d+)?)', expand=False)
#converting data type from object to int
hiv_main_data['Number of deaths due to HIV/AIDS']=hiv_main_data['Number of deaths due to HIV/AIDS'].astype(int)
```

```python
hiv_main_data['New HIV infections(per 1000 uninfected pop).(Male)']=hiv_main_data['New HIV infections(per 1000 uninfected pop).(Male)'].astype(float)
```

```python
#separating 0.01 from &lt;0.01 in the col named Number of deaths due to HIV/AIDS.
hiv_main_data['New HIV infections(per 1000 uninfected pop).(Both_sexes)']=hiv_main_data['New HIV infections(per 1000 uninfected pop).(Both_sexes)'].str.extract(r'(\d+(?:\.\d+)?)', expand=False)
#converting data type from object to int
hiv_main_data['New HIV infections(per 1000 uninfected pop).(Both_sexes)']=hiv_main_data['New HIV infections(per 1000 uninfected pop).(Both_sexes)'].astype(float)
```

```python
#separating 0.01 from &lt;0.01 in the col named Number of deaths due to HIV/AIDS.
hiv_main_data['New HIV infections(per 1000 uninfected pop).(Female)']=hiv_main_data['New HIV infections(per 1000 uninfected pop).(Female)'].str.extract(r'(\d+(?:\.\d+)?)', expand=False)
#converting data type from object to int
hiv_main_data['New HIV infections(per 1000 uninfected pop).(Female)']=hiv_main_data['New HIV infections(per 1000 uninfected pop).(Female)'].astype(float)
```

```python
hiv_main_data['PEOPLE HAVING HIV']=hiv_main_data['PEOPLE HAVING HIV'].astype(int)
```
